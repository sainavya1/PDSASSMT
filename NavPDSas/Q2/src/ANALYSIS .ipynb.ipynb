```python
import matplotlib.pyplot as plt    
import seaborn as sb              
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


data = pd.read_csv('StudentsPerformance.csv')

data['total score'] = data['math score'] + data['reading score'] + data['writing score']
data['average score'] = data['total score'] / 3
data.head()
#displaying dataframe
data
```

```python
plt.rcParams['figure.figsize'] = (20, 6)

# Create a subplot with bar charts
plt.subplot(1, 4, 1)
ax =sb.histplot(data, x = 'gender', color='violet')

# Continue with similar code blocks for other pie charts
plt.subplot(1, 4, 2)
ax =sb.histplot(data, x = 'race/ethnicity', color='red')


plt.subplot(1, 4, 3)
ax =sb.histplot(data, x = 'parental level of education',color='purple')
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')

plt.subplot(1, 4, 4)
ax =sb.histplot(data, x = 'lunch', color='black')

plt.grid()
plt.show()
```

```python

sb.barplot(data=data,x='race/ethnicity',y='math score',hue='gender')
```

```python
plt.subplots(1,3,figsize=(20,20))

plt.subplot(131)
ax =sb.histplot(data=data,x='average score',hue='lunch')

plt.subplot(132)
ax =sb.histplot(data=data[data.gender=='male'],x='average score',hue='lunch')

plt.subplot(133)
ax =sb.histplot(data=data[data.gender=='female'],x='average score',hue='lunch')
plt.show()

```

```python
sb.violinplot(data,x="parental level of education",y="average score",hue='gender', split=True)
```

```python
sb.catplot(data=data,x='gender',y='reading score',kind='box',col='test preparation course', whis=[0,100])
plt.show()
```
